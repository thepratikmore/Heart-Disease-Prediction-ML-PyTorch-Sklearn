## Heart Disease Prediction Using PyTorch
Step 3: Model Evaluation

```python
# Import Required Libraries
import torch
import torch.nn as nn
```

```python
# Load Test Data
X_test = torch.tensor(torch.load("X_test.pt"), dtype=torch.float32)
y_test = torch.tensor(torch.load("y_test.pt"), dtype=torch.float32).view(-1, 1)
```

```python
# Recreate Model Architecture
class HeartDiseaseModel(nn.Module):
    
    def __init__(self):
        super().__init__()
        self.layer1 = nn.Linear(13, 16)
        self.layer2 = nn.Linear(16, 8)
        self.output = nn.Linear(8, 1)
        self.relu = nn.ReLU()
        self.sigmoid = nn.Sigmoid()

    def forward(self, x):
        x = self.relu(self.layer1(x))
        x = self.relu(self.layer2(x))
        x = self.sigmoid(self.output(x))
        return x
```

```python
# Load Saved Model Weights
model = HeartDiseaseModel()
model.load_state_dict(torch.load("heart_model.pth"))
model.eval()
```

```python
# Calculate Accuracy
with torch.no_grad():predictions = model(X_test)

# Convert probabilities to 0 or 1
predictions = (predictions > 0.5).float()
accuracy = (predictions == y_test).sum() / len(y_test)

print(f"Model Accuracy: {accuracy.item() * 100:.2f}%")
```
